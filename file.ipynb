```python
name = input("Enter your name: ")
age = input("Enter your age: ")

with open("user_data.txt", "w") as f:
    f.write(f"Name: {name}\nAge: {age}")

with open("user_data.txt", "r") as f:
    print("\nStored Data:")
    print(f.read())
```

```python
import csv

# Step 1: Write to a CSV file
with open("students.csv", mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Age", "Grade"])
    writer.writerow(["Alice", 14, "A"])
    writer.writerow(["Bob", 15, "B+"])
    writer.writerow(["Charlie", 13, "A-"])

# Step 2: Read from the CSV file
with open("students.csv", mode="r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
# Open source image in binary read mode
with open("cmpt.jpg", "rb") as src_file:
    data = src_file.read()

# Write the binary data to a new file
with open("copy_image.jpg", "wb") as dest_file:
    dest_file.write(data)
    print(data)

print("Image copied successfully.")
```

```python
# Write a program to read a file and count:
# Total characters
# Total words
# Total lines


file = open('user_data.txt', 'r')
lines = file.readlines()
print(lines)
file.close()

total_lines = len(lines)
total_words = 0
total_chars = 0

for line in lines:
    total_chars += len(line)
    total_words += len(line.split())

print("Total Characters:", total_chars)
print("Total Words:", total_words)
print("Total Lines:", total_lines)
```
